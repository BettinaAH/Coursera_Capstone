# Next step: transformation of data to be easier for analysis by One Hot Encoding
Therefor I will put the different conditions of the columns like "weather" or "collisiontype" into different columns.

```python
df_an = df_capstone
```

```python
#drop more unnecessary columns
df_an = df_an.drop(['OBJECTID','SEVERITYDESC','INCDATE' , 'INCDTTM','INCKEY', 'COLDETKEY', 'REPORTNO', 'LOCATION','STATUS', 'SEGLANEKEY', 'CROSSWALKKEY', 'SDOT_COLDESC','ST_COLDESC', 'HITPARKEDCAR','X','Y','SEVERITYCODE.1'],axis=1)
```

```python
# splitting the column of weather
df_an[pd.get_dummies(df_an['WEATHER']).columns] = pd.get_dummies(df_an['WEATHER'])
df_an.drop(['WEATHER'], axis=1, inplace=True)

# splitting the column of collisiontype
df_an[pd.get_dummies(df_an['COLLISIONTYPE']).columns] = pd.get_dummies(df_an['COLLISIONTYPE'])
df_an.drop(['COLLISIONTYPE'], axis=1, inplace=True)

# splitting the column of Addresstype 
df_an[pd.get_dummies(df_an['ADDRTYPE']).columns] = pd.get_dummies(df_an['ADDRTYPE'])
df_an.drop(['ADDRTYPE'], axis=1, inplace=True)

# splitting the column of junctiontype
df_an[pd.get_dummies(df_an['JUNCTIONTYPE']).columns] = pd.get_dummies(df_an['JUNCTIONTYPE'])
df_an.drop(['JUNCTIONTYPE'], axis=1, inplace=True)

# splitting the column of roadconditions
df_an[pd.get_dummies(df_an['ROADCOND']).columns] = pd.get_dummies(df_an['ROADCOND'])
df_an.drop(['ROADCOND'], axis=1, inplace=True)

# splitting the column of lightconditions
df_an[pd.get_dummies(df_an['LIGHTCOND']).columns] = pd.get_dummies(df_an['LIGHTCOND'])
df_an.drop(['LIGHTCOND'], axis=1, inplace=True)

df_an.head()
```

```python
# In the column Inattiontion there seems to be two ways of labels (yes (Y) / no (N) and 0 and 1), so I will align that and only make it 0 or 1

df_an['INATTENTIONIND'] = df_an['INATTENTIONIND'].map({'Y': '1', 'N': '0'})
```

```python
df_an['INATTENTIONIND'] = df_an['INATTENTIONIND'] .astype('int64')
```

# Starting with modeling phase where various algorithms and methods are tried out

```python
# looking at the correlation between the different attributes

df_an.corr().style.background_gradient(cmap='coolwarm').format("{:.2f}")
```

## What are we learning from the correlation table?
It is acutally really suprising, that there are so little correlation with the **severity of the accident**. All attributes show a very low correlation. 

The ones with a little higher correlation (corr > 0.2) are accidents at intersections or accistents that involved parked cars, pedestrians or cyclists. Which would mean if you as a driver approach a intersection, parked cars, pedestrians or cyclists: be more careful. Also the ones responsible for designing the intersections, pedestrian crossings, cycle lanes and parking lots should be mindful in design them.

But when I started this projects, I assumpted that bad weather would be a big influence of probability of an accident.
It does not look like it when looking at this big data set of Seattle.

# Looking deeper into the ones that are slightly more correlated

```python
import sklearn.model_selection as model_selection
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
# setting up for training and train/test split

df_modeling = df_an[['SEVERITYCODE','PERSONCOUNT','PEDCOUNT','PEDCYLCOUNT','Parked Car','PEDROWNOTGRNT','Intersection', 'Cycles', 'Pedestrian', 'At Intersection (intersection related)']]
X = df_modeling.drop(['SEVERITYCODE'],axis=1)
y = df_modeling['SEVERITYCODE']
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, train_size=0.8,test_size=0.2)

print("done")
```

```python
# Looking at the Logistic Regression

lr = LogisticRegression()
lr.fit(X_train,y_train)
yhat_lr=lr.predict(X_test)

print("done")
```

```python
# looking into Support Vector Machine

clf = svm.SVC(kernel = "linear")
clf.fit(X_train, y_train)
yhat_svm = clf.predict(X_test)
print("done")
```

```python
# looking at the Decision Tree

dt = DecisionTreeClassifier(criterion = "entropy", max_depth=10)
dt.fit(X_train, y_train)
yhat_dt = dt.predict(X_test)

print("done")
```

```python
# looking into Random Forest Classifier

m = RandomForestClassifier(n_estimators=100, criterion="entropy")
m.fit(X_train, y_train)
yhat_rfc = m.predict(X_test)

print("done")
```

```python
# Overview of different scores from the modeling
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
f1lr = f1_score(y_test, yhat_lr)
f1dt = f1_score(y_test, yhat_dt)
f1rf = f1_score(y_test, yhat_rfc)
f1svm = f1_score(y_test, yhat_svm)


acclr = accuracy_score(y_test, yhat_lr)
accdt = accuracy_score(y_test, yhat_dt)
accrf = accuracy_score(y_test, yhat_rfc)
accsvm = accuracy_score(y_test, yhat_svm)


plr = precision_score(y_test, yhat_lr)
pdt = precision_score(y_test, yhat_dt)
prf = precision_score(y_test, yhat_rfc)
psvm = precision_score(y_test, yhat_svm)

rlr = recall_score(y_test, yhat_lr)
rdt = recall_score(y_test, yhat_dt)
rrf = recall_score(y_test, yhat_rfc)
rsvm = recall_score(y_test, yhat_svm)


results = [['Model', 'Accuracy score', 'F1 score', 'Precision','Recall'],['Logistic Regression', acclr, f1lr, plr, rlr],['Decision Tree',accdt, f1dt,pdt,rdt],['Random Forest',accrf,f1rf,prf,rrf],['Support Vector Machine', accsvm, f1svm, psvm, rsvm]]
results = pd.DataFrame(results)       
print(results)
```

# One more look into the road and weather conditions

what if I add up the bad road and weather conditions to "good" and "bad". As it might not matter if the road is icy or wet, it both might increase the propability of an accident.

That's why I will adjust the data so that there are only two conditions good = 0 and bad = 1

```python
df_01 = df_capstone
```

```python
df_01.shape
```

## deleting rows where the Roadcondition is marked as "other" or "unknown" or "oil"

```python
# replace "Other" and "Unknown" to NaN
df_01.replace("Other", np.nan, inplace = True)
df_01.replace("Unknown", np.nan, inplace = True)
df_01.replace("Oil", np.nan, inplace = True) # shortcut to delete also Roadconditions with "oil", as they are not weather dependent

# drop whole row with NaN in "ROADCOND" column
df_01.dropna(subset=["ROADCOND"], axis=0, inplace=True)

# drop whole row with NaN in "WEATHER" column
df_01.dropna(subset=["WEATHER"], axis=0, inplace=True)


df_01.shape
```

```python
# Replacing ROADCOND values with 0 for good/fine weather and 1 for bad weather
df_01['ROADCOND'].replace(to_replace=['Wet','Dry','Snow/Slush','Ice','Sand/Mud/Dirt','Standing Water'], value = ['1','0','1','1','1','1'], inplace=True)
```

```python
# Replacing Weather values with 0 for good/fine weather and 1 for bad weather
df_01['WEATHER'].replace(to_replace=['Clear','Raining','Overcast','Snowing','Fog/Smog/Smoke','Sleet/Hail/Freezing Rain','Blowing Sand/Dirt','Severe Crosswind','Partly Cloudy'], value = ['0','1','0','1','1','1','1','1','0'], inplace=True)
```

```python
df_SRW = df_01[['SEVERITYCODE','ROADCOND','WEATHER']] # SRW stands for Serveritycode Roadconditions Weather
df_SRW.head()
```

## Looking at Correlation of Severitycode and the grouped bad/fine weather columns Roadcondition and Weather
